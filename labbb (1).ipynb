```python

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
cat_feature = ["prognosis"]
onehot = OneHotEncoder(handle_unknown="ignore")
transformer = ColumnTransformer([("onehot", onehot, cat_feature)], remainder = "passthrough")
encoded_x = transformer.fit_transform(x_train)
encoded_x_test = transformer.transform(x_test)
encoded_x
```

"accuracy_score(y_test, predictions)" calculates the accuracy by comparing the true labels (y_test) with the predicted labels (predictions).

The accuracy is then printed as a percentage.

Precision: The ratio of correctly predicted positive observations to the total predicted positives.

Recall (Sensitivity): The ratio of correctly predicted positive observations to all observations in the actual class.

F1-Score: The weighted average of Precision and Recall.

Support: The number of actual occurrences of each class in the specified dataset.
